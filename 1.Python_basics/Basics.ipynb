Задание 2
Дана переменная, в которой хранится четырехзначное число (год). 
Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Если год не делится на 4, значит он обычный.
Иначе надо проверить не делится ли год на 100.
Если не делится, значит это не столетие и можно сделать вывод, что год високосный.
Если делится на 100, значит это столетие и его следует проверить его делимость на 400.
Если год делится на 400, то он високосный.
Иначе год обычный.

```python
year = 2021
if year % 4 != 0:
    print('Обычный год')
elif year % 100 == 0: 
        #Столетие
        if  year % 400 == 0:
           print('Високосный год') 
        else:
           print('Обычный год')
else:
    print('Високосный год')
```

Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет “счастливым”. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

```python
#Не работает с лидирующими нулями
number = 123456
n1_1 = number // 100000
n1_2 = (number-n1_1*100000) // 10000
n1_3 = (number-n1_1*100000-n1_2*10000) // 1000
n2_1 = (number-n1_1*100000-n1_2*10000 - n1_3*1000) // 100
n2_2 = (number-n1_1*100000-n1_2*10000 - n1_3*1000 - n2_1*100) // 10
n2_3 = number-n1_1*100000-n1_2*10000 - n1_3*1000 - n2_1*100 - n2_2*10
if n1_1 + n1_2 + n1_3 == n2_1 + n2_2 + n2_3:
    print('Счастливый билет')
else:
    print('Несчастливый билет')
```

Задание 6 (необязательное)
Напишите программу, которая сможет вычислять площади трех фигур (круг, треугольник и прямоугольник). 
Тип фигуры запрашиваем через пользовательский ввод, после чего делаем запрос характеристик фигуры:

если пользователь выбрал круг, запрашиваем его радиус,
если треугольник – длины трех его сторон;
если прямоугольник – длины двух его сторон.

```python

figure_type = input('Введите тип фигуры(круг, треугольник и прямоугольник):')
correct_type = True
if figure_type == "круг":
    r = float(input('Введите радиус:'))
    sqr = 3.14*(r**2)
elif figure_type == "треугольник":
    a = float(input('Введите длину стороны а:'))
    b = float(input('Введите длину стороны b:'))
    c = float(input('Введите длину стороны c:'))
    p = (a + b + c)/2
    sqr = (p*(p-a)*(p-b)*(p-c))**(.5)
elif figure_type == "прямоугольник":
    a = float(input('Введите длину стороны а:'))
    b = float(input('Введите длину стороны b:'))
    sqr = a*b
else:
    correct_type = False

if correct_type == True:
    print('Площадь фигуры: %f' % sqr)
else:
    print('тип фигуры введен некорректно')

    
```
